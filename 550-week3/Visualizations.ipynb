```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gni = pd.read_csv("gni_totals.csv")
```

```python
gni.columns = ['Country','GNI']
```

```python
plt.barh(gni.Country, gni.GNI)
plt.title('Middle East GNI in 2017') 
plt.ylabel('Country')
plt.xlabel('GNI')
plt.xticks(rotation=90)
```

```python
plt.plot(gni.Country, gni.GNI)
plt.title('Middle East GNI in 2017')
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('GNI')
plt.show()
```

```python
plt.scatter(gni.Country,gni.GNI)
plt.title('Middle East GNI in 2017')
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('GNI')
plt.show()
```

The most effective visualization, in my opinion, is the horizontal bar chart. The second most effective is the scatterplot, and the least effective would be the line chart. I believe that the horizontal bar chart is the most effective because it not only gives you the GNI per country, but it also gives you a sense of how that GNI compares to the other countries simply by looking at the length of each bar. In other words, if I were to remove all the numbers, you would be able to compare the differences in proportion of all the countries to see roughly how the GNI differs among them. I would say that the scatterplot is the second most effective, because first and foremost it shows the lienar relationship between the countries and their respective GNIs. It also does give  bit of an idea as to the proportionality between the countries, by looking at how much distance lies between each point. The least effective would be the line chart. With the line chart, it is barely even possible to see which countries have which GNIs, which is not something that should be estimated. It also isn't visually appealing, nor conclusive of anything, besides the liner relationship that exists.
